Estadisticas de oro por equipo

```python
s10.columns[s10.columns.str.contains("gold")]
```

```python
gold_stats = s10[s10.columns[s10.columns.str.contains("gold")]]

#lado azul

blue_gold_stats = gold_stats[gold_stats.columns[gold_stats.columns.str.contains("blue")]]

#total
s10["blue_totalgold"] = blue_gold_stats[blue_gold_stats.columns[blue_gold_stats.columns.str.contains("total")]].sum(axis=1)

#earned
s10["blue_earnedgold"] = blue_gold_stats[blue_gold_stats.columns[blue_gold_stats.columns.str.contains("earned")]].sum(axis=1)

#spent
s10["blue_spentgold"] = blue_gold_stats[blue_gold_stats.columns[blue_gold_stats.columns.str.contains("spent")]].sum(axis=1)

#at10
s10["blue_goldat10"] = blue_gold_stats[blue_gold_stats.columns[blue_gold_stats.columns.str.contains("at10")]].sum(axis=1)


#at15
s10["blue_goldat15"] = blue_gold_stats[blue_gold_stats.columns[blue_gold_stats.columns.str.contains("at15")]].sum(axis=1)
```

```python
red_gold_stats = gold_stats[gold_stats.columns[gold_stats.columns.str.contains("red")]]

#total
s10["red_totalgold"] = red_gold_stats[red_gold_stats.columns[red_gold_stats.columns.str.contains("total")]].sum(axis=1)

#earned
s10["red_earnedgold"] = red_gold_stats[red_gold_stats.columns[red_gold_stats.columns.str.contains("earned")]].sum(axis=1)

#spent
s10["red_spentgold"] = red_gold_stats[red_gold_stats.columns[red_gold_stats.columns.str.contains("spent")]].sum(axis=1)

#at10
s10["red_goldat10"] = red_gold_stats[red_gold_stats.columns[red_gold_stats.columns.str.contains("at10")]].sum(axis=1)


#at15
s10["red_goldat15"] = red_gold_stats[red_gold_stats.columns[red_gold_stats.columns.str.contains("at15")]].sum(axis=1)
```

```python
cols_card = []
card = []

for i in s10:
    cols_card.append(i)
    card.append((s10[i].nunique()/len(s10)*100))

p100_card = pd.DataFrame(index=cols_card, data=card)
p100_card
```

```python
card_0 = pd.concat([cardi,p100_card], axis=1, )
card_0
```
