## Cardinalidad

```python
unique = s10.nunique()
unique
```

```python
for i in s10:
    print(f"La cardinalidad de la variable {i} es {(s10[i].nunique()/s10.shape[0])*100:.2f}%")
```

```python
cols_card = []
card = []

for i in s10:
    cols_card.append(i)
    card.append((s10[i].nunique()/s10.shape[0])*100)

p100_card = pd.DataFrame(index=cols_card, data=card)
p100_card
```

```python
card_0 = pd.concat([unique,p100_card], axis=1, )
card_0
```

```python
tipo_datos = s10.dtypes
tipo_datos
```

```python
cardinalidad = pd.concat([card_0, tipo_datos], axis=1)
cardinalidad.columns = ["uniques", "cardinalidad", "data_type"]
```
