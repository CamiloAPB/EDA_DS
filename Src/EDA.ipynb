### Hipótesis

La variable objetivo era el resultado por lo tanto las tres hipótesis se plantearan en torno a esta. 

La primera hipótesis se basaba en lo observado en los gráficos, en los que el equipo azul tendía a ganar más. Para comprobar esta hipótesis realizomos una prueba t para solo una muestra. 

La segunda hipótesis tenía en cuenta la influencia del oro acumulado en el resultado. Primero verificamos la correlación entre las variables de oro total de cada equipo y el resultado para decidir que prueba realizar. Tras comprobar que las variables estaban significatibamente correlacionadas decidimos realizar una prueba *Wilcoxon* para cada equipo,  ya que estas variables siguen la distribución normal.

La tercera hipótesis busca comprobar si la obtención de los tres primeros objetivos mas importantes del early game (primer heraldo, primer dragón y primera torre de medio) no influyen realmente en el resultado como vimos en los mapas de calor. Como en la hipótesis anterior, verificamos la correlación de las variables. Al verificar que están correlacionadas y saber que no siguen la distrbución normal se decide realizar una prueba de *Friedman*.

Por último realizamos una prueba de *Shappiro-Wilks* para cada variable numérica para confirmar que no siguen una distribución normal.
